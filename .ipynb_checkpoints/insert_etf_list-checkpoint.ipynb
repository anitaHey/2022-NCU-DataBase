```python
import pymysql
import requests
from bs4 import BeautifulSoup


db_settings = {
    "host": "127.0.0.1",
    "port": 3306,
    "user": "root",
    "password": "",
    "db": "ncu_database",
    "charset": "utf8"
}
```

```python
def getHTML():
    response = requests.get("https://www.twse.com.tw/zh/ETF/list")
    soup = BeautifulSoup(response.text, "html.parser")
    
    result = soup.select("table tr")
    for tr in result:
        all_td = tr.select('td')
        if(len(all_td) > 0):
            stock_id = str(all_td[1]).split("<br/>")
            stock_name = str(all_td[2]).split("<br/>")
            
            for index in range(len(stock_id)):
                print(BeautifulSoup(stock_id[index]).getText() + " : " + BeautifulSoup(stock_name[index]).getText())
    
getHTML()
```

```python
try:
    conn = pymysql.connect(**db_settings)
    with conn.cursor() as cursor:
        command = "INSERT INTO etf_list (id, name) VALUES (%s, %s)"
        
        response = requests.get("https://www.twse.com.tw/zh/ETF/list")
        soup = BeautifulSoup(response.text, "html.parser")

        result = soup.select("table tr")
        for tr in result:
            all_td = tr.select('td')
            if(len(all_td) > 0):
                stock_id = str(all_td[1]).split("<br/>")
                stock_name = str(all_td[2]).split("<br/>")

                for index in range(len(stock_id)):
                    cursor.execute(command, (BeautifulSoup(stock_id[index]).getText(), BeautifulSoup(stock_name[index]).getText()))
                    
        conn.commit()
except Exception as ex:
    print(ex)
```
