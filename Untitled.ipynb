```python
import mplfinance as mpf
import pymysql
import json
import pandas as pd

db_settings = {
    "host": "127.0.0.1",
    "port": 3306,
    "user": "root",
    "password": "",
    "db": "ncu_database",
    "charset": "utf8"
}
```

```python
try:  
    conn = pymysql.connect(**db_settings)
    with conn.cursor() as cursor:
        command = "SELECT d, o, h, l, tz, tv FROM stock_data WHERE d like '110/12/%' and c = '0050' "
        cursor.execute(command)
        conn.commit()
        result = cursor.fetchall()
except Exception as ex:
    print(ex)
```

```python
arr = []
for r in result:
    r = list(r)
    r[0] = r[0].replace('110', '2021', 1);
    r[1] = float(r[1])
    r[2] = float(r[2])
    r[3] = float(r[3])
    r[4] = float(r[4])
    r[5] = float(r[5])
    arr.append(r)
```

```python
arr_df = pd.DataFrame(arr)
arr_df.index = pd.to_datetime(arr_df[0])
arr_df = arr_df.drop(columns=[0])
arr_df.columns = ['Open', 'High', 'Low', 'Close', 'Volume']
arr_df.index.name = "Date"
arr_df
```

```python
mpf.plot(arr_df, type='candle', style='charles')
```
